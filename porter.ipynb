# Train Test Split

```python
df[['market_id', 'store_primary_category', 'order_protocol', 'created_at_hour', 'created_at_day']] = df[['market_id', 'store_primary_category', 'order_protocol', 'created_at_hour', 'created_at_day']].astype('category')
```

```python
df.drop(columns=['created_at', 'actual_delivery_time'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train_val , X_test, y_train_val, y_test = train_test_split(df.drop(columns=['delivery_time']),df['delivery_time'], test_size=0.2, random_state=42)

X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
X_train.info()
```

# Encoding

```python
from category_encoders  import TargetEncoder

encoder = TargetEncoder(cols=['market_id', 'store_primary_category', 'order_protocol', 'created_at_hour', 'created_at_day'])

encoder.fit(X_train, y_train)

X_train_encoded = encoder.transform(X_train)
X_val_encoded = encoder.transform(X_val)
X_test_encoded = encoder.transform(X_test)
```

```python
X_train_encoded = pd.DataFrame(X_train_encoded, columns=X_train.columns)
X_val_encoded = pd.DataFrame(X_val_encoded, columns=X_val.columns)
X_test_encoded = pd.DataFrame(X_test_encoded, columns=X_test.columns)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train_encoded), columns=X_train_encoded.columns)
X_val_scaled = pd.DataFrame(scaler.transform(X_val_encoded), columns=X_val_encoded.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test_encoded), columns=X_test_encoded.columns)
X_test_scaled.head()
```

# NN Training

```python
import tensorflow as tf
from tensorflow.keras import Sequential, regularizers
from tensorflow.keras.layers import Input, Dense, BatchNormalization, Dropout
```

```python
X_train_scaled.shape[1]
```

```python
model = Sequential([
    Input(shape=(X_train_scaled.shape[1],)),    
    Dense(64, activation='relu', kernel_initializer='glorot_normal'),        
    Dense(128, activation='relu', kernel_initializer='glorot_normal'),    
    Dense(128, activation='relu', kernel_initializer='glorot_normal'),    
    Dense(128, activation='relu', kernel_initializer='glorot_normal'),    
    Dense(64, activation='relu', kernel_initializer='glorot_normal'),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', 
              loss='mse', 
              metrics=['mse', 'mape', 'r2_score'])
```

```python
history = model.fit(X_train_scaled, y_train, validation_data=(X_val_scaled, y_val), epochs=25, batch_size=256)
```

```python
epochs = history.epoch
loss = history.history["loss"]
val_loss = history.history["val_loss"]


plt.plot(epochs, loss, label="train")
plt.plot(epochs, val_loss, label="validation")

plt.legend()
plt.title("Loss VS Epochs")

plt.show()
```

## Hyperparameter tuning

```python
import keras_tuner
```

```python
np.arange(0.01, 0.22,0.05)
```

```python
class MyNN(keras_tuner.HyperModel):
    def build(self,hp):
        model = Sequential([Input(shape=(X_train_scaled.shape[1],))])
        for i in range(1, hp.Int('num_layers', min_value=1, max_value=5)):
            if hp.Boolean("batch_norm"):
                model.add(BatchNormalization())
            if hp.Boolean("dropout"):
                model.add(Dropout(rate=hp.Choice('dropout_rate', values=[0.1,0.2,0.3])))
            model.add(
                Dense(
                    name=f'dense_layer_{i}',
                    units=hp.Int('units', min_value=32, max_value=129, step=32), 
                    activation='relu', 
                    kernel_initializer='glorot_normal',
                    # kernel_regularizer=hp.Float('kernal_regu', min_value=0.01, max_value=0.22, step=0.05)
                )
            )
        model.add(Dense(1))

        hp_learning_rate = hp.Choice("learning_rate", values=[1e-2, 1e-3, 1e-4])
        
        # Define optimizer, loss, and metrics
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate),
                    loss=tf.keras.losses.MeanSquaredError(),
                    metrics=["r2_score", "mse", "mape"])
        
        return model

    def fit(self, hp, model, *args, **kwargs):
        return model.fit(
            *args,
            batch_size=hp.Int("batch_size", min_value=64, max_value=257, step=64),            
            **kwargs,
        )
```

```python
MyNN(keras_tuner.HyperParameters())
```

```python
tuner = keras_tuner.RandomSearch(
    MyNN(),
    objective="mape",        
    overwrite=True,
    directory="nn_tuning",
    project_name="helloworld",    
)
```

```python
tuner.search(X_train_scaled, y_train, validation_data=(X_val_scaled, y_val))
```

```python
class MyNNGS(keras_tuner.HyperModel):
    def build(self,hp):
        model = Sequential([Input(shape=(X_train_scaled.shape[1],))])
        for i in range(1, 5):            
            model.add(
                Dense(
                    name=f'dense_layer_{i}',
                    units=hp.Int('units', min_value=64, max_value=129, step=32), 
                    activation='relu', 
                    kernel_initializer='glorot_normal',                    
                )
            )
        model.add(Dense(1))

        hp_learning_rate = hp.Choice("learning_rate", values=[1e-2, 2e-2, 3e-2])
        
        # Define optimizer, loss, and metrics
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate),
                    loss=tf.keras.losses.MeanSquaredError(),
                    metrics=["r2_score", "mse", "mape"])
        
        return model

    def fit(self, hp, model, *args, **kwargs):
        return model.fit(
            *args,
            batch_size=hp.Int("batch_size", min_value=128, max_value=257, step=64),            
            **kwargs,
        )
```
